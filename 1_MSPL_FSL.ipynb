**Multi-Space Prototypical Learning (MSPL) with FGSM**

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix
from collections import defaultdict
from tabulate import tabulate
import random
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.metrics import cohen_kappa_score

# Set global font sizes for all plots
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['legend.fontsize'] = 14

# Load and preprocess data (unchanged)
def load_and_preprocess_data(filepath):
    data = pd.read_csv(filepath)
    X = data.iloc[:, :-1].values
    y = data.iloc[:, -1].values

    # Encode labels
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)
    class_names = label_encoder.classes_
    num_classes = len(class_names)

    # Standardize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    return X_scaled, y_encoded, class_names, num_classes

# Split data into meta-train and meta-test sets (unchanged)
def split_meta_data(y_encoded, num_classes, test_size=0.3, random_state=42):
    np.random.seed(random_state)
    all_classes = np.unique(y_encoded)
    meta_train_classes = np.random.choice(
        all_classes, int((1 - test_size) * num_classes), replace=False)
    meta_test_classes = np.array([c for c in all_classes if c not in meta_train_classes])

    train_mask = np.isin(y_encoded, meta_train_classes)
    test_mask = np.isin(y_encoded, meta_test_classes)

    return train_mask, test_mask

# Few-shot sampling function (unchanged)
def sample_episode(X, y, n_way, k_shot, q_query, return_classes=False, force_classes=None):
    unique_classes = np.unique(y)

    if force_classes is not None:
        selected_classes = force_classes
        n_way = len(selected_classes)
    else:
        n_way = min(n_way, len(unique_classes))
        if len(unique_classes) < n_way:
            raise ValueError(f"Requested {n_way}-way classification but only {len(unique_classes)} classes available")
        selected_classes = np.random.choice(unique_classes, n_way, replace=False)

    support, support_labels, query, query_labels = [], [], [], []

    for idx, cls in enumerate(selected_classes):
        cls_indices = np.where(y == cls)[0]
        if len(cls_indices) < k_shot + q_query:
            q_query_adjusted = len(cls_indices) - k_shot
            if q_query_adjusted <= 0:
                support_samples = np.random.choice(cls_indices, k_shot, replace=True)
                query_samples = np.array([], dtype=int)
            else:
                support_samples = np.random.choice(cls_indices, k_shot, replace=False)
                query_samples = np.random.choice(
                    np.setdiff1d(cls_indices, support_samples),
                    q_query_adjusted,
                    replace=False
                )
        else:
            support_samples = np.random.choice(cls_indices, k_shot, replace=False)
            remaining = np.setdiff1d(cls_indices, support_samples)
            query_samples = np.random.choice(remaining, q_query, replace=False)

        support.append(X[support_samples])
        support_labels.append([idx] * len(support_samples))
        if len(query_samples) > 0:
            query.append(X[query_samples])
            query_labels.append([idx] * len(query_samples))

    support = np.vstack(support) if len(support) > 0 else np.empty((0, X.shape[1]))
    query = np.vstack(query) if len(query) > 0 else np.empty((0, X.shape[1]))
    support_labels = np.hstack(support_labels) if len(support_labels) > 0 else np.array([], dtype=int)
    query_labels = np.hstack(query_labels) if len(query_labels) > 0 else np.array([], dtype=int)

    if return_classes:
        return (
            torch.FloatTensor(support),
            torch.LongTensor(support_labels),
            torch.FloatTensor(query),
            torch.LongTensor(query_labels),
            selected_classes
        )
    return (
        torch.FloatTensor(support),
        torch.LongTensor(support_labels),
        torch.FloatTensor(query),
        torch.LongTensor(query_labels)
    )

# Multi-Space Prototypical Learning (MSPL) Network
class MSPLNet(nn.Module):
    def __init__(self, input_dim, hidden_dim=256, embed_dim=128, num_spaces=3):
        super().__init__()
        self.num_spaces = num_spaces
        self.embed_dim = embed_dim

        # Shared feature extractor
        self.shared_net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.BatchNorm1d(hidden_dim)
        )

        # Multiple projection spaces
        self.projection_heads = nn.ModuleList([
            nn.Sequential(
                nn.Linear(hidden_dim, embed_dim),
                nn.ReLU()
            ) for _ in range(num_spaces)
        ])

        # Space importance weights
        self.space_weights = nn.Parameter(torch.ones(num_spaces) / num_spaces)

        # Adversarial defense components
        self.epsilon = 0.1  # Perturbation magnitude for FGSM
        self.adv_coeff = 0.5  # Adversarial loss coefficient

    def forward(self, x, return_all=False):
        shared_features = self.shared_net(x)
        projections = [head(shared_features) for head in self.projection_heads]

        if return_all:
            return projections
        else:
            # Combine projections using learned weights
            weights = torch.softmax(self.space_weights, dim=0)
            combined = sum(w * p for w, p in zip(weights, projections))
            return combined

    def fgsm_attack(self, x, y):
        """Generate FGSM adversarial examples"""
        x_adv = x.clone().detach().requires_grad_(True)

        # Forward pass
        embeddings = self(x_adv)

        # For FGSM in few-shot setting, we'll use the prototype loss
        with torch.enable_grad():
            # Compute prototypes in each space
            all_prototypes = []
            unique_classes_in_episode = torch.unique(y)
            # Map original episode labels to a contiguous range for loss calculation
            label_map = {cls.item(): i for i, cls in enumerate(unique_classes_in_episode)}
            y_mapped = torch.LongTensor([label_map[label.item()] for label in y])

            # Get projections from all spaces
            space_projections = self(x_adv, return_all=True)

            # Compute loss for each space and combine
            total_loss = 0
            for proj in space_projections:
                prototypes = []
                for cls in unique_classes_in_episode:
                    class_embeddings = proj[y == cls]
                    if len(class_embeddings) > 0:
                        prototypes.append(class_embeddings.mean(dim=0))
                prototypes = torch.stack(prototypes) if len(prototypes) > 0 else torch.zeros((len(unique_classes_in_episode), self.embed_dim))

                # Compute loss for this space
                dists = torch.cdist(proj, prototypes)
                logits = -dists
                total_loss += nn.CrossEntropyLoss()(logits, y_mapped)

            # Compute gradient
            total_loss.backward()

            # Get gradient sign
            perturbation = self.epsilon * x_adv.grad.sign()
            x_adv = x_adv + perturbation

        return x_adv.detach()

    def adversarial_loss(self, clean_x, adv_x, y):
        """Compute adversarial loss term using KL divergence across all spaces"""
        clean_embs = self(clean_x, return_all=True)
        adv_embs = self(adv_x, return_all=True)

        total_kl_loss = 0

        for clean_emb, adv_emb in zip(clean_embs, adv_embs):
            # Compute prototypes from clean embeddings
            prototypes = []
            for cls in torch.unique(y):
                prototypes.append(clean_emb[y == cls].mean(dim=0))
            prototypes = torch.stack(prototypes)

            # Compute both clean and adversarial distances
            clean_dists = torch.cdist(clean_emb, prototypes)
            adv_dists = torch.cdist(adv_emb, prototypes)

            # KL divergence between clean and adversarial distributions
            clean_probs = torch.softmax(-clean_dists, dim=1) + 1e-9
            adv_probs = torch.softmax(-adv_dists, dim=1) + 1e-9
            kl_loss = nn.KLDivLoss(reduction='batchmean')(torch.log(adv_probs), clean_probs)
            total_kl_loss += kl_loss

        return total_kl_loss / len(clean_embs)  # Average KL loss across spaces

# Training loop for MSPL
def train_mspl(model, optimizer, X_train, y_train, X_val, y_val,
              n_way, k_shot, q_query, epochs=25, episodes_per_epoch=100,
              adversarial_training=True):
    model.train()
    best_val_acc = 0
    history = {'train_loss': [], 'train_acc': [], 'val_acc': []}

    for epoch in range(epochs):
        epoch_loss, epoch_acc = 0.0, 0.0
        epoch_adv_loss = 0.0

        # Training phase
        model.train()
        for _ in range(episodes_per_epoch):
            # Sample clean episode
            support_x, support_y, query_x, query_y = sample_episode(
                X_train, y_train, n_way, k_shot, q_query)

            # Generate FGSM adversarial examples if using adversarial training
            if adversarial_training:
                with torch.enable_grad():
                    adv_support_x = model.fgsm_attack(support_x, support_y)
                    adv_query_x = model.fgsm_attack(query_x, query_y)

            # Forward pass for clean data - get projections from all spaces
            space_support_embs = model(support_x, return_all=True)
            space_query_embs = model(query_x, return_all=True)

            # Compute loss for each space and combine
            clean_loss = 0
            space_acc = 0

            for support_emb, query_emb in zip(space_support_embs, space_query_embs):
                # Compute prototypes for this space
                prototypes = []
                for cls in torch.unique(support_y):
                    prototypes.append(support_emb[support_y == cls].mean(dim=0))
                prototypes = torch.stack(prototypes)

                # Compute clean loss for this space
                dists = torch.cdist(query_emb, prototypes)
                logits = -dists
                clean_loss += nn.CrossEntropyLoss()(logits, query_y)
                space_acc += (logits.argmax(dim=1) == query_y).float().mean().item()

            clean_loss /= len(space_support_embs)  # Average loss across spaces
            space_acc /= len(space_support_embs)   # Average accuracy across spaces

            # Compute adversarial loss if using adversarial training
            if adversarial_training:
                # Compute adversarial loss term
                adv_loss = model.adversarial_loss(support_x, adv_support_x, support_y)
                adv_loss += model.adversarial_loss(query_x, adv_query_x, query_y)

                # Combine losses
                total_loss = clean_loss + model.adv_coeff * adv_loss
                epoch_adv_loss += adv_loss.item()
            else:
                total_loss = clean_loss

            optimizer.zero_grad()
            total_loss.backward()
            optimizer.step()

            epoch_loss += total_loss.item()
            epoch_acc += space_acc  # Use average space accuracy

        avg_train_loss = epoch_loss / episodes_per_epoch
        avg_train_acc = epoch_acc / episodes_per_epoch
        avg_adv_loss = epoch_adv_loss / episodes_per_epoch if adversarial_training else 0

        # Validation phase
        val_acc = evaluate_mspl(model, X_val, y_val, n_way, k_shot, q_query)

        # Save best model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), 'best_model.pth')

        # Record history
        history['train_loss'].append(avg_train_loss)
        history['train_acc'].append(avg_train_acc)
        history['val_acc'].append(val_acc)

        if adversarial_training:
            print(f"Epoch {epoch+1}/{epochs} | Loss: {avg_train_loss:.4f} (Clean Last: {clean_loss.item():.4f}, Adv Avg: {avg_adv_loss:.4f}) | "
                  f"Train Acc: {avg_train_acc:.4f} | Val Acc: {val_acc:.4f}")
        else:
            print(f"Epoch {epoch+1}/{epochs} | Loss: {avg_train_loss:.4f} | "
                  f"Train Acc: {avg_train_acc:.4f} | Val Acc: {val_acc:.4f}")

    # Load best model
    model.load_state_dict(torch.load('best_model.pth'))
    return history

# Evaluation helper function for MSPL
def evaluate_mspl(model, X, y, n_way, k_shot, q_query, num_episodes=50):
    model.eval()
    val_acc = 0

    for _ in range(num_episodes):
        support_x, support_y, query_x, query_y = sample_episode(X, y, n_way, k_shot, q_query)
        with torch.no_grad():
            # Get projections from all spaces
            space_support_embs = model(support_x, return_all=True)
            space_query_embs = model(query_x, return_all=True)

            # Combine predictions from all spaces
            combined_logits = None

            for support_emb, query_emb in zip(space_support_embs, space_query_embs):
                # Compute prototypes for this space
                prototypes = []
                for cls in torch.unique(support_y):
                    prototypes.append(support_emb[support_y == cls].mean(dim=0))
                prototypes = torch.stack(prototypes)

                # Compute logits for this space
                dists = torch.cdist(query_emb, prototypes)
                space_logits = -dists

                if combined_logits is None:
                    combined_logits = space_logits
                else:
                    combined_logits += space_logits

            # Average logits across spaces
            combined_logits /= len(space_support_embs)
            val_acc += (combined_logits.argmax(dim=1) == query_y).float().mean().item()

    return val_acc / num_episodes

def compute_metrics_from_confusion(conf_matrix):
    """Compute metrics from a confusion matrix"""
    diag = np.diag(conf_matrix)
    row_sum = conf_matrix.sum(axis=1)  # True counts per class (TP + FN)
    col_sum = conf_matrix.sum(axis=0)  # Predicted counts per class (TP + FP)
    total = conf_matrix.sum()

    # Handle division by zero
    precision = np.zeros_like(diag, dtype=float)
    recall = np.zeros_like(diag, dtype=float)
    f1 = np.zeros_like(diag, dtype=float)

    for i in range(len(diag)):
        if col_sum[i] > 0:
            precision[i] = diag[i] / col_sum[i]
        if row_sum[i] > 0:
            recall[i] = diag[i] / row_sum[i]
        if (precision[i] + recall[i]) > 0:
            f1[i] = 2 * (precision[i] * recall[i]) / (precision[i] + recall[i])

    # Calculate aggregated metrics
    overall_accuracy = np.sum(diag) / total if total > 0 else 0


    # Weighted averages
    weighted_precision = np.sum(precision * row_sum) / total if total > 0 else 0
    weighted_recall = np.sum(recall * row_sum) / total if total > 0 else 0
    weighted_f1 = np.sum(f1 * row_sum) / total if total > 0 else 0

    return {
        'accuracy': overall_accuracy,
        'weighted_precision': weighted_precision,
        'weighted_recall': weighted_recall,
        'weighted_f1': weighted_f1,
        'class_precision': precision,
        'class_recall': recall,
        'class_f1': f1
    }


# Enhanced comprehensive evaluation with additional metrics (modified for MSPL)
def enhanced_comprehensive_evaluate(model, X_test, y_test, class_names, num_classes, n_way, k_shot, q_query,
                                  num_episodes=100, attack_strength=0.1):
    """Enhanced evaluation with additional metrics and visualizations"""
    model.eval()
    # Initialize confusion matrices with total number of classes
    clean_conf_matrix = np.zeros((num_classes, num_classes), dtype=int)
    adv_conf_matrix = np.zeros((num_classes, num_classes), dtype=int)

    all_true_labels_global = []
    all_clean_pred_labels_global = []
    all_adv_pred_labels_global = []
    clean_confidences = []
    adv_confidences = []

    for _ in tqdm(range(num_episodes), desc="Evaluating"):
        try:
            support_x, support_y, query_x, query_y, selected_classes = sample_episode(
                X_test, y_test, n_way, k_shot, q_query, return_classes=True)

            if len(query_y) == 0:
                continue

            # Clean evaluation
            with torch.no_grad():
                # Get projections from all spaces
                space_support_embs = model(support_x, return_all=True)
                space_query_embs = model(query_x, return_all=True)

                # Combine predictions from all spaces
                combined_clean_logits = None

                for support_emb, query_emb in zip(space_support_embs, space_query_embs):
                    # Compute prototypes for this space
                    prototypes = []
                    for cls in torch.unique(support_y):
                        prototypes.append(support_emb[support_y == cls].mean(dim=0))
                    prototypes = torch.stack(prototypes)

                    # Compute logits for this space
                    dists = torch.cdist(query_emb, prototypes)
                    space_logits = -dists

                    if combined_clean_logits is None:
                        combined_clean_logits = space_logits
                    else:
                        combined_clean_logits += space_logits

                # Average logits across spaces
                combined_clean_logits /= len(space_support_embs)
                clean_probs = torch.softmax(combined_clean_logits, dim=1)
                clean_preds = combined_clean_logits.argmax(dim=1).cpu().numpy()
                clean_max_probs = clean_probs.max(dim=1)[0].cpu().numpy()
                true_labels_episode = query_y.cpu().numpy()

            # Generate FGSM adversarial examples
            query_x_adv = query_x.clone().detach().requires_grad_(True)

            # Compute loss for gradient using all spaces
            space_adv_embs_attack = model(query_x_adv, return_all=True)

            # Compute total loss across all spaces
            total_loss = 0
            for adv_emb in space_adv_embs_attack:
                # Need to compute prototypes based on the support set of the current episode
                prototypes = []
                for cls in torch.unique(support_y):
                    class_embeddings = model(support_x)[support_y == cls]
                    if len(class_embeddings) > 0:
                        prototypes.append(class_embeddings.mean(dim=0))
                prototypes = torch.stack(prototypes) if len(prototypes) > 0 else torch.zeros((len(torch.unique(support_y)), model.embed_dim))

                adv_logits_attack = -torch.cdist(adv_emb, prototypes)
                total_loss += nn.CrossEntropyLoss()(adv_logits_attack, query_y)

            total_loss /= len(space_adv_embs_attack)  # Average loss across spaces

            # Compute gradient
            grad = torch.autograd.grad(total_loss, query_x_adv)[0]

            # Apply FGSM perturbation
            adv_query_x = query_x_adv + attack_strength * grad.sign()
            adv_query_x = adv_query_x.detach()

            # Evaluate adversarial examples
            with torch.no_grad():
                # Get projections from all spaces for adversarial examples
                space_adv_embs_eval = model(adv_query_x, return_all=True)

                # Combine predictions from all spaces
                combined_adv_logits = None

                for adv_emb in space_adv_embs_eval:
                    # Use the same prototypes from the clean support set for adversarial evaluation
                    prototypes = []
                    for cls in torch.unique(support_y):
                        class_embeddings = model(support_x)[support_y == cls]
                        if len(class_embeddings) > 0:
                            prototypes.append(class_embeddings.mean(dim=0))
                    prototypes = torch.stack(prototypes)

                    # Compute logits for this space
                    dists = torch.cdist(adv_emb, prototypes)
                    space_logits = -dists

                    if combined_adv_logits is None:
                        combined_adv_logits = space_logits
                    else:
                        combined_adv_logits += space_logits

                # Average logits across spaces
                combined_adv_logits /= len(space_adv_embs_eval)
                adv_probs = torch.softmax(combined_adv_logits, dim=1)
                adv_preds = combined_adv_logits.argmax(dim=1).cpu().numpy()
                adv_max_probs = adv_probs.max(dim=1)[0].cpu().numpy()

            # Map episode labels to global class indices and collect
            true_global_labels_episode = [selected_classes[label] for label in true_labels_episode]
            clean_pred_global_labels_episode = [selected_classes[label] for label in clean_preds]
            adv_pred_global_labels_episode = [selected_classes[label] for label in adv_preds]

            all_true_labels_global.extend(true_global_labels_episode)
            all_clean_pred_labels_global.extend(clean_pred_global_labels_episode)
            all_adv_pred_labels_global.extend(adv_pred_global_labels_episode)
            clean_confidences.extend(clean_max_probs)
            adv_confidences.extend(adv_max_probs)

            # Update confusion matrices
            for true, pred in zip(true_global_labels_episode, clean_pred_global_labels_episode):
                clean_conf_matrix[true, pred] += 1
            for true, pred in zip(true_global_labels_episode, adv_pred_global_labels_episode):
                adv_conf_matrix[true, pred] += 1

        except ValueError as e:
            print(f"Skipping episode: {e}")
            continue

    # Calculate metrics from confusion matrices
    clean_metrics = compute_metrics_from_confusion(clean_conf_matrix)
    adv_metrics = compute_metrics_from_confusion(adv_conf_matrix)

    # Calculate additional robustness metrics
    clean_correct = np.sum(np.diag(clean_conf_matrix))
    adv_correct = np.sum(np.diag(adv_conf_matrix))
    total_samples = np.sum(clean_conf_matrix)

    # Attack Success Rate (ASR)
    if clean_correct > 0:
        asr = (clean_correct - adv_correct) / clean_correct
    else:
        asr = 0.0

    # Robust Accuracy Ratio
    if clean_metrics['accuracy'] > 0:
        robust_ratio = adv_metrics['accuracy'] / clean_metrics['accuracy']
    else:
        robust_ratio = 0.0

    # Mean confidence drop
    mean_confidence_drop = np.mean(np.array(clean_confidences) - np.array(adv_confidences)) if clean_confidences else 0

    # Cohen's Kappa
    kappa_clean = cohen_kappa_score(all_true_labels_global, all_clean_pred_labels_global)
    kappa_adv = cohen_kappa_score(all_true_labels_global, all_adv_pred_labels_global)

    # Add additional metrics
    clean_metrics.update({
        'kappa': kappa_clean,
        'mean_confidence': np.mean(clean_confidences) if clean_confidences else 0
    })

    adv_metrics.update({
        'attack_success_rate': asr,
        'robust_accuracy_ratio': robust_ratio,
        'mean_confidence_drop': mean_confidence_drop,
        'kappa': kappa_adv,
        'mean_confidence': np.mean(adv_confidences) if adv_confidences else 0
    })

    return clean_metrics, adv_metrics, (clean_conf_matrix, adv_conf_matrix)

def plot_robustness_curves(results):
    """Plot clean vs adversarial accuracy across attack strengths with enhanced fonts"""
    # Filter results to get unique attack strengths and corresponding accuracies
    unique_strengths = sorted(list(set([r['attack_strength'] for r in results])))
    standard_clean_accs = []
    standard_adv_accs = []
    robust_clean_accs = []
    robust_adv_accs = []

    for strength in unique_strengths:
        # Get all scenarios for this attack strength and average them
        scenario_data = [r for r in results if r['attack_strength'] == strength]
        standard_clean_accs.append(np.mean([r['standard_clean_acc'] for r in scenario_data]))
        standard_adv_accs.append(np.mean([r['standard_adv_acc'] for r in scenario_data]))
        robust_clean_accs.append(np.mean([r['robust_clean_acc'] for r in scenario_data]))
        robust_adv_accs.append(np.mean([r['robust_adv_acc'] for r in scenario_data]))

    plt.figure(figsize=(10, 6))
    plt.plot(unique_strengths, standard_clean_accs, 'b-', label='Standard (Clean)', linewidth=2)
    plt.plot(unique_strengths, standard_adv_accs, 'b--', label='Standard (Adv)', linewidth=2)
    plt.plot(unique_strengths, robust_clean_accs, 'g-', label='Robust (Clean)', linewidth=2)
    plt.plot(unique_strengths, robust_adv_accs, 'g--', label='Robust (Adv)', linewidth=2)

    plt.xlabel('Attack Strength (ε)', fontsize=14)
    plt.ylabel('Accuracy', fontsize=14)
    plt.title('Model Robustness Across Attack Strengths', fontsize=16)
    plt.legend(fontsize=14)
    plt.grid(True)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.show()

def plot_performance_vs_shots(model, X, y, n_way, q_query, max_shots=20, num_trials=10, attack_strength=0.1):
    """Plot accuracy vs number of support shots for different few-shot scenarios"""
    shot_counts = range(1, max_shots+1)

    # We'll evaluate three scenarios: 1-shot, 5-shot, and 10-shot
    scenarios = [
        {'n_way': n_way, 'k_shot': 1, 'label': '1-shot', 'color': 'b', 'marker': 'o'},
        {'n_way': n_way, 'k_shot': 5, 'label': '5-shot', 'color': 'g', 'marker': 's'},
        {'n_way': n_way, 'k_shot': 10, 'label': '10-shot', 'color': 'r', 'marker': 'd'}
    ]

    plt.figure(figsize=(10, 6))

    for scenario in scenarios:
        accuracies = []
        for k in shot_counts:
            acc = 0
            for _ in range(num_trials):
                acc += evaluate_mspl(model, X, y, scenario['n_way'], k, q_query)
            accuracies.append(acc / num_trials)

        plt.plot(shot_counts, accuracies,
                 label=f"{scenario['label']}",
                 color=scenario['color'],
                 marker=scenario['marker'],
                 linewidth=1)

    plt.xlabel('Number of Support Shots (k)', fontsize=16)
    plt.ylabel('Accuracy', fontsize=16)
    plt.title(f'Few-Shot Performance vs Support Set Size', fontsize=16)
    plt.legend(fontsize=14)
    plt.grid(True)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.show()

def print_detailed_metrics(metrics, class_names, title):
    """Print detailed metrics including per-class precision, recall, and F1"""
    print(f"\n{title} Metrics:")
    print(f"Accuracy: {metrics['accuracy']:.4f}")
    print(f"Weighted Precision: {metrics['weighted_precision']:.4f}")
    print(f"Weighted Recall: {metrics['weighted_recall']:.4f}")
    print(f"Weighted F1: {metrics['weighted_f1']:.4f}")

    if 'kappa' in metrics:
        print(f"Cohen's Kappa: {metrics['kappa']:.4f}")

    if 'attack_success_rate' in metrics:
        print(f"Attack Success Rate: {metrics['attack_success_rate']:.4f}")
        print(f"Robust Accuracy Ratio: {metrics['robust_accuracy_ratio']:.4f}")
        print(f"Mean Confidence Drop: {metrics['mean_confidence_drop']:.4f}")

# Modified scenario testing function to use MSPL
def run_adversarial_scenarios(X_train, y_train, X_test, y_test, class_names, num_classes, scenarios):
    results = []

    for scenario in scenarios:
        n_way = scenario['n_way']
        k_shot = scenario['k_shot']
        q_query = scenario.get('q_query', 15)
        attack_strength = scenario.get('attack_strength', 0.1)

        print(f"\nRunning scenario: {n_way}-way {k_shot}-shot with FGSM attack strength {attack_strength}")

        # Split train into train and validation
        X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
            X_train, y_train, test_size=0.2, random_state=42)

        # Initialize standard model (no adversarial training)
        standard_model = MSPLNet(X_train.shape[1])
        standard_optimizer = optim.Adam(standard_model.parameters(), lr=0.001)

        # Train standard model
        print("Training standard MSPL model...")
        train_mspl(
            standard_model, standard_optimizer, X_train_split, y_train_split,
            X_val_split, y_val_split, n_way, k_shot, q_query, epochs=25,
            episodes_per_epoch=100, adversarial_training=False)

        # Initialize robust model (with adversarial training)
        robust_model = MSPLNet(X_train.shape[1])
        robust_optimizer = optim.Adam(robust_model.parameters(), lr=0.001)

        # Train robust model
        print("Training robust MSPL model with FGSM adversarial training...")
        train_mspl(
            robust_model, robust_optimizer, X_train_split, y_train_split,
            X_val_split, y_val_split, n_way, k_shot, q_query, epochs=25,
            episodes_per_epoch=100, adversarial_training=True)

        # Evaluate both models under FGSM attack with enhanced metrics
        print("Evaluating standard MSPL model under FGSM attack...")
        standard_clean_metrics, standard_adv_metrics, standard_confs = enhanced_comprehensive_evaluate(
            standard_model, X_test, y_test, class_names, num_classes, n_way, k_shot, q_query,
            num_episodes=100, attack_strength=attack_strength)

        print("Evaluating robust MSPL model under FGSM attack...")
        robust_clean_metrics, robust_adv_metrics, robust_confs = enhanced_comprehensive_evaluate(
            robust_model, X_test, y_test, class_names, num_classes, n_way, k_shot, q_query,
            num_episodes=100, attack_strength=attack_strength)

        # Print detailed metrics
        print_detailed_metrics(standard_clean_metrics, class_names, "Standard MSPL Model (Clean)")
        print_detailed_metrics(standard_adv_metrics, class_names, "Standard MSPL Model (Adversarial)")
        print_detailed_metrics(robust_clean_metrics, class_names, "Robust MSPL Model (Clean)")
        print_detailed_metrics(robust_adv_metrics, class_names, "Robust MSPL Model (Adversarial)")

        # Generate visualizations for robust model
        print("\nGenerating visualizations for robust model...")

        # Only plot performance vs support size for attack strength 0.1
        if attack_strength == 0.1:
            plot_performance_vs_shots(robust_model, X_test, y_test, n_way, q_query, attack_strength=attack_strength)

        # Store results
        scenario_result = {
            'n_way': n_way,
            'k_shot': k_shot,
            'attack_strength': attack_strength,
            'standard_clean_acc': standard_clean_metrics['accuracy'],
            'standard_adv_acc': standard_adv_metrics['accuracy'],
            'standard_drop': standard_clean_metrics['accuracy'] - standard_adv_metrics['accuracy'],
            'robust_clean_acc': robust_clean_metrics['accuracy'],
            'robust_adv_acc': robust_adv_metrics['accuracy'],
            'robust_drop': robust_clean_metrics['accuracy'] - robust_adv_metrics['accuracy'],
            'improvement': (robust_adv_metrics['accuracy'] - standard_adv_metrics['accuracy']) / standard_adv_metrics['accuracy'] if standard_adv_metrics['accuracy'] > 0 else 0,
            'standard_asr': standard_adv_metrics['attack_success_rate'],
            'robust_asr': robust_adv_metrics['attack_success_rate'],
        }
        results.append(scenario_result)

    # Plot robustness curves after all scenarios
    plot_robustness_curves(results)

    return results

# Main execution with enhanced evaluation
if __name__ == "__main__":
    # Load and preprocess data
    X_scaled, y_encoded, class_names, num_classes = load_and_preprocess_data(
        '/content/drive/MyDrive/ID2024-5G/Preprocessed_5G.csv')

    # Split into meta-train and meta-test sets
    train_mask, test_mask = split_meta_data(y_encoded, num_classes)
    X_train, y_train = X_scaled[train_mask], y_encoded[train_mask]
    X_test, y_test = X_scaled[test_mask], y_encoded[test_mask]

    # Define scenarios to test with different FGSM attack strengths
    scenarios = [
        {'n_way': 5, 'k_shot': 1, 'attack_strength': 0.05},
        {'n_way': 5, 'k_shot': 5, 'attack_strength': 0.05},
        {'n_way': 5, 'k_shot': 10, 'attack_strength': 0.05},

        {'n_way': 5, 'k_shot': 1, 'attack_strength': 0.10},
        {'n_way': 5, 'k_shot': 5, 'attack_strength': 0.10},
        {'n_way': 5, 'k_shot': 10, 'attack_strength': 0.10},

        {'n_way': 5, 'k_shot': 1, 'attack_strength': 0.15},
        {'n_way': 5, 'k_shot': 5, 'attack_strength': 0.15},
        {'n_way': 5, 'k_shot': 10, 'attack_strength': 0.15},

        {'n_way': 5, 'k_shot': 1, 'attack_strength': 0.20},
        {'n_way': 5, 'k_shot': 5, 'attack_strength': 0.20},
        {'n_way': 5, 'k_shot': 10, 'attack_strength': 0.20},
    ]

    # Run enhanced adversarial scenarios with MSPL
    results = run_adversarial_scenarios(
        X_train, y_train, X_test, y_test, class_names, num_classes, scenarios)

    # Display results
    print("\nEnhanced Adversarial Scenario Results:")
    print(tabulate(results, headers="keys", tablefmt="grid", floatfmt=".4f"))
```

**Multi-Space Prototypical Learning (MSPL) with PGD**

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix
from collections import defaultdict
from tabulate import tabulate
import random
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.metrics import cohen_kappa_score

# Set global font sizes for all plots
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['legend.fontsize'] = 14

# Load and preprocess data (unchanged)
def load_and_preprocess_data(filepath):
    data = pd.read_csv(filepath)
    X = data.iloc[:, :-1].values
    y = data.iloc[:, -1].values

    # Encode labels
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)
    class_names = label_encoder.classes_
    num_classes = len(class_names)

    # Standardize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    return X_scaled, y_encoded, class_names, num_classes

# Split data into meta-train and meta-test sets (unchanged)
def split_meta_data(y_encoded, num_classes, test_size=0.3, random_state=42):
    np.random.seed(random_state)
    all_classes = np.unique(y_encoded)
    meta_train_classes = np.random.choice(
        all_classes, int((1 - test_size) * num_classes), replace=False)
    meta_test_classes = np.array([c for c in all_classes if c not in meta_train_classes])

    train_mask = np.isin(y_encoded, meta_train_classes)
    test_mask = np.isin(y_encoded, meta_test_classes)

    return train_mask, test_mask

# Few-shot sampling function (unchanged)
def sample_episode(X, y, n_way, k_shot, q_query, return_classes=False, force_classes=None):
    unique_classes = np.unique(y)

    if force_classes is not None:
        selected_classes = force_classes
        n_way = len(selected_classes)
    else:
        n_way = min(n_way, len(unique_classes))
        if len(unique_classes) < n_way:
            raise ValueError(f"Requested {n_way}-way classification but only {len(unique_classes)} classes available")
        selected_classes = np.random.choice(unique_classes, n_way, replace=False)

    support, support_labels, query, query_labels = [], [], [], []

    for idx, cls in enumerate(selected_classes):
        cls_indices = np.where(y == cls)[0]
        if len(cls_indices) < k_shot + q_query:
            q_query_adjusted = len(cls_indices) - k_shot
            if q_query_adjusted <= 0:
                support_samples = np.random.choice(cls_indices, k_shot, replace=True)
                query_samples = np.array([], dtype=int)
            else:
                support_samples = np.random.choice(cls_indices, k_shot, replace=False)
                query_samples = np.random.choice(
                    np.setdiff1d(cls_indices, support_samples),
                    q_query_adjusted,
                    replace=False
                )
        else:
            support_samples = np.random.choice(cls_indices, k_shot, replace=False)
            remaining = np.setdiff1d(cls_indices, support_samples)
            query_samples = np.random.choice(remaining, q_query, replace=False)

        support.append(X[support_samples])
        support_labels.append([idx] * len(support_samples))
        if len(query_samples) > 0:
            query.append(X[query_samples])
            query_labels.append([idx] * len(query_samples))

    support = np.vstack(support) if len(support) > 0 else np.empty((0, X.shape[1]))
    query = np.vstack(query) if len(query) > 0 else np.empty((0, X.shape[1]))
    support_labels = np.hstack(support_labels) if len(support_labels) > 0 else np.array([], dtype=int)
    query_labels = np.hstack(query_labels) if len(query_labels) > 0 else np.array([], dtype=int)

    if return_classes:
        return (
            torch.FloatTensor(support),
            torch.LongTensor(support_labels),
            torch.FloatTensor(query),
            torch.LongTensor(query_labels),
            selected_classes
        )
    return (
        torch.FloatTensor(support),
        torch.LongTensor(support_labels),
        torch.FloatTensor(query),
        torch.LongTensor(query_labels)
    )

# Multi-Space Prototypical Learning (MSPL) Network with PGD attack
class MSPLNet(nn.Module):
    def __init__(self, input_dim, hidden_dim=256, embed_dim=128, num_spaces=3):
        super().__init__()
        self.num_spaces = num_spaces
        self.embed_dim = embed_dim

        # Shared feature extractor
        self.shared_net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.BatchNorm1d(hidden_dim)
        )

        # Multiple projection spaces
        self.projection_heads = nn.ModuleList([
            nn.Sequential(
                nn.Linear(hidden_dim, embed_dim),
                nn.ReLU()
            ) for _ in range(num_spaces)
        ])

        # Space importance weights
        self.space_weights = nn.Parameter(torch.ones(num_spaces) / num_spaces)

        # Adversarial defense components
        self.epsilon = 0.1  # Maximum perturbation magnitude
        self.adv_coeff = 0.5  # Adversarial loss coefficient
        self.pgd_steps = 7  # Number of PGD steps
        self.pgd_step_size = 0.03  # Step size for PGD

    def forward(self, x, return_all=False):
        shared_features = self.shared_net(x)
        projections = [head(shared_features) for head in self.projection_heads]

        if return_all:
            return projections
        else:
            # Combine projections using learned weights
            weights = torch.softmax(self.space_weights, dim=0)
            combined = sum(w * p for w, p in zip(weights, projections))
            return combined

    def pgd_attack(self, x, y):
        """Generate PGD adversarial examples"""
        x_adv = x.clone().detach()

        # Random start within the epsilon ball
        x_adv = x_adv + torch.empty_like(x_adv).uniform_(-self.epsilon, self.epsilon)
        x_adv = torch.clamp(x_adv, x.min(), x.max()).detach()

        for _ in range(self.pgd_steps):
            x_adv.requires_grad_(True)

            # Forward pass through all spaces
            space_projections = self(x_adv, return_all=True)

            # Compute loss for each space and combine
            total_loss = 0
            unique_classes_in_episode = torch.unique(y)
            # Map original episode labels to a contiguous range for loss calculation
            label_map = {cls.item(): i for i, cls in enumerate(unique_classes_in_episode)}
            y_mapped = torch.LongTensor([label_map[label.item()] for label in y])

            for proj in space_projections:
                # Compute prototypes for this space
                prototypes = []
                for cls in unique_classes_in_episode:
                    class_embeddings = proj[y == cls]
                    if len(class_embeddings) > 0:
                        prototypes.append(class_embeddings.mean(dim=0))
                prototypes = torch.stack(prototypes) if len(prototypes) > 0 else torch.zeros((len(unique_classes_in_episode), self.embed_dim))

                # Compute loss for this space
                dists = torch.cdist(proj, prototypes)
                logits = -dists
                total_loss += nn.CrossEntropyLoss()(logits, y_mapped)

            # Average loss across spaces
            total_loss /= len(space_projections)

            # Compute gradient
            grad = torch.autograd.grad(total_loss, x_adv)[0]

            # Update adversarial examples with gradient sign
            x_adv = x_adv.detach() + self.pgd_step_size * grad.sign()

            # Project back to epsilon ball and valid data range
            delta = torch.clamp(x_adv - x, -self.epsilon, self.epsilon)
            x_adv = torch.clamp(x + delta, x.min(), x.max()).detach()

        return x_adv

    def adversarial_loss(self, clean_x, adv_x, y):
        """Compute adversarial loss term using KL divergence across all spaces"""
        clean_embs = self(clean_x, return_all=True)
        adv_embs = self(adv_x, return_all=True)

        total_kl_loss = 0

        for clean_emb, adv_emb in zip(clean_embs, adv_embs):
            # Compute prototypes from clean embeddings
            prototypes = []
            for cls in torch.unique(y):
                prototypes.append(clean_emb[y == cls].mean(dim=0))
            prototypes = torch.stack(prototypes)

            # Compute both clean and adversarial distances
            clean_dists = torch.cdist(clean_emb, prototypes)
            adv_dists = torch.cdist(adv_emb, prototypes)

            # KL divergence between clean and adversarial distributions
            clean_probs = torch.softmax(-clean_dists, dim=1) + 1e-9
            adv_probs = torch.softmax(-adv_dists, dim=1) + 1e-9
            kl_loss = nn.KLDivLoss(reduction='batchmean')(torch.log(adv_probs), clean_probs)
            total_kl_loss += kl_loss

        return total_kl_loss / len(clean_embs)  # Average KL loss across spaces

# Training loop for MSPL with PGD
def train_mspl(model, optimizer, X_train, y_train, X_val, y_val,
              n_way, k_shot, q_query, epochs=25, episodes_per_epoch=100,
              adversarial_training=True):
    model.train()
    best_val_acc = 0
    history = {'train_loss': [], 'train_acc': [], 'val_acc': []}

    for epoch in range(epochs):
        epoch_loss, epoch_acc = 0.0, 0.0
        epoch_adv_loss = 0.0

        # Training phase
        model.train()
        for _ in range(episodes_per_epoch):
            # Sample clean episode
            support_x, support_y, query_x, query_y = sample_episode(
                X_train, y_train, n_way, k_shot, q_query)

            # Generate PGD adversarial examples if using adversarial training
            if adversarial_training:
                with torch.enable_grad():
                    adv_support_x = model.pgd_attack(support_x, support_y)
                    adv_query_x = model.pgd_attack(query_x, query_y)

            # Forward pass for clean data - get projections from all spaces
            space_support_embs = model(support_x, return_all=True)
            space_query_embs = model(query_x, return_all=True)

            # Compute loss for each space and combine
            clean_loss = 0
            space_acc = 0

            for support_emb, query_emb in zip(space_support_embs, space_query_embs):
                # Compute prototypes for this space
                prototypes = []
                for cls in torch.unique(support_y):
                    prototypes.append(support_emb[support_y == cls].mean(dim=0))
                prototypes = torch.stack(prototypes)

                # Compute clean loss for this space
                dists = torch.cdist(query_emb, prototypes)
                logits = -dists
                clean_loss += nn.CrossEntropyLoss()(logits, query_y)
                space_acc += (logits.argmax(dim=1) == query_y).float().mean().item()

            clean_loss /= len(space_support_embs)  # Average loss across spaces
            space_acc /= len(space_support_embs)   # Average accuracy across spaces

            # Compute adversarial loss if using adversarial training
            if adversarial_training:
                # Compute adversarial loss term
                adv_loss = model.adversarial_loss(support_x, adv_support_x, support_y)
                adv_loss += model.adversarial_loss(query_x, adv_query_x, query_y)

                # Combine losses
                total_loss = clean_loss + model.adv_coeff * adv_loss
                epoch_adv_loss += adv_loss.item()
            else:
                total_loss = clean_loss

            optimizer.zero_grad()
            total_loss.backward()
            optimizer.step()

            epoch_loss += total_loss.item()
            epoch_acc += space_acc  # Use average space accuracy

        avg_train_loss = epoch_loss / episodes_per_epoch
        avg_train_acc = epoch_acc / episodes_per_epoch
        avg_adv_loss = epoch_adv_loss / episodes_per_epoch if adversarial_training else 0

        # Validation phase
        val_acc = evaluate_mspl(model, X_val, y_val, n_way, k_shot, q_query)

        # Save best model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), 'best_model.pth')

        # Record history
        history['train_loss'].append(avg_train_loss)
        history['train_acc'].append(avg_train_acc)
        history['val_acc'].append(val_acc)

        if adversarial_training:
            print(f"Epoch {epoch+1}/{epochs} | Loss: {avg_train_loss:.4f} (Clean Last: {clean_loss.item():.4f}, Adv Avg: {avg_adv_loss:.4f}) | "
                  f"Train Acc: {avg_train_acc:.4f} | Val Acc: {val_acc:.4f}")
        else:
            print(f"Epoch {epoch+1}/{epochs} | Loss: {avg_train_loss:.4f} | "
                  f"Train Acc: {avg_train_acc:.4f} | Val Acc: {val_acc:.4f}")

    # Load best model
    model.load_state_dict(torch.load('best_model.pth'))
    return history

# Evaluation helper function for MSPL (unchanged)
def evaluate_mspl(model, X, y, n_way, k_shot, q_query, num_episodes=50):
    model.eval()
    val_acc = 0

    for _ in range(num_episodes):
        support_x, support_y, query_x, query_y = sample_episode(X, y, n_way, k_shot, q_query)
        with torch.no_grad():
            # Get projections from all spaces
            space_support_embs = model(support_x, return_all=True)
            space_query_embs = model(query_x, return_all=True)

            # Combine predictions from all spaces
            combined_logits = None

            for support_emb, query_emb in zip(space_support_embs, space_query_embs):
                # Compute prototypes for this space
                prototypes = []
                for cls in torch.unique(support_y):
                    prototypes.append(support_emb[support_y == cls].mean(dim=0))
                prototypes = torch.stack(prototypes)

                # Compute logits for this space
                dists = torch.cdist(query_emb, prototypes)
                space_logits = -dists

                if combined_logits is None:
                    combined_logits = space_logits
                else:
                    combined_logits += space_logits

            # Average logits across spaces
            combined_logits /= len(space_support_embs)
            val_acc += (combined_logits.argmax(dim=1) == query_y).float().mean().item()

    return val_acc / num_episodes

def compute_metrics_from_confusion(conf_matrix):
    """Compute metrics from a confusion matrix"""
    diag = np.diag(conf_matrix)
    row_sum = conf_matrix.sum(axis=1)  # True counts per class (TP + FN)
    col_sum = conf_matrix.sum(axis=0)  # Predicted counts per class (TP + FP)
    total = conf_matrix.sum()

    # Handle division by zero
    precision = np.zeros_like(diag, dtype=float)
    recall = np.zeros_like(diag, dtype=float)
    f1 = np.zeros_like(diag, dtype=float)

    for i in range(len(diag)):
        if col_sum[i] > 0:
            precision[i] = diag[i] / col_sum[i]
        if row_sum[i] > 0:
            recall[i] = diag[i] / row_sum[i]
        if (precision[i] + recall[i]) > 0:
            f1[i] = 2 * (precision[i] * recall[i]) / (precision[i] + recall[i])

    # Calculate aggregated metrics
    overall_accuracy = np.sum(diag) / total if total > 0 else 0

    # Weighted averages
    weighted_precision = np.sum(precision * row_sum) / total if total > 0 else 0
    weighted_recall = np.sum(recall * row_sum) / total if total > 0 else 0
    weighted_f1 = np.sum(f1 * row_sum) / total if total > 0 else 0

    return {
        'accuracy': overall_accuracy,
        'weighted_precision': weighted_precision,
        'weighted_recall': weighted_recall,
        'weighted_f1': weighted_f1,
        'class_precision': precision,
        'class_recall': recall,
        'class_f1': f1
    }

# Enhanced comprehensive evaluation with PGD attack
def enhanced_comprehensive_evaluate(model, X_test, y_test, class_names, num_classes, n_way, k_shot, q_query,
                                  num_episodes=100, attack_strength=0.1):
    """Enhanced evaluation with additional metrics and visualizations"""
    model.eval()
    # Initialize confusion matrices with total number of classes
    clean_conf_matrix = np.zeros((num_classes, num_classes), dtype=int)
    adv_conf_matrix = np.zeros((num_classes, num_classes), dtype=int)

    all_true_labels_global = []
    all_clean_pred_labels_global = []
    all_adv_pred_labels_global = []
    clean_confidences = []
    adv_confidences = []

    for _ in tqdm(range(num_episodes), desc="Evaluating"):
        try:
            support_x, support_y, query_x, query_y, selected_classes = sample_episode(
                X_test, y_test, n_way, k_shot, q_query, return_classes=True)

            if len(query_y) == 0:
                continue

            # Clean evaluation
            with torch.no_grad():
                # Get projections from all spaces
                space_support_embs = model(support_x, return_all=True)
                space_query_embs = model(query_x, return_all=True)

                # Combine predictions from all spaces
                combined_clean_logits = None

                for support_emb, query_emb in zip(space_support_embs, space_query_embs):
                    # Compute prototypes for this space
                    prototypes = []
                    for cls in torch.unique(support_y):
                        prototypes.append(support_emb[support_y == cls].mean(dim=0))
                    prototypes = torch.stack(prototypes)

                    # Compute logits for this space
                    dists = torch.cdist(query_emb, prototypes)
                    space_logits = -dists

                    if combined_clean_logits is None:
                        combined_clean_logits = space_logits
                    else:
                        combined_clean_logits += space_logits

                # Average logits across spaces
                combined_clean_logits /= len(space_support_embs)
                clean_probs = torch.softmax(combined_clean_logits, dim=1)
                clean_preds = combined_clean_logits.argmax(dim=1).cpu().numpy()
                clean_max_probs = clean_probs.max(dim=1)[0].cpu().numpy()
                true_labels_episode = query_y.cpu().numpy()

            # Generate PGD adversarial examples
            query_x_adv = query_x.clone().detach()

            # Random start within the epsilon ball
            query_x_adv = query_x_adv + torch.empty_like(query_x_adv).uniform_(-attack_strength, attack_strength)
            query_x_adv = torch.clamp(query_x_adv, query_x.min(), query_x.max()).detach()

            for _ in range(model.pgd_steps):
                query_x_adv.requires_grad_(True)

                # Forward pass through all spaces
                space_adv_embs_attack = model(query_x_adv, return_all=True)

                # Compute loss for each space and combine
                total_loss = 0
                unique_classes_in_episode = torch.unique(support_y)
                # Map original episode labels to a contiguous range for loss calculation
                label_map = {cls.item(): i for i, cls in enumerate(unique_classes_in_episode)}
                y_mapped = torch.LongTensor([label_map[label.item()] for label in query_y])

                for adv_emb in space_adv_embs_attack:
                    # Need to compute prototypes based on the support set of the current episode
                    prototypes = []
                    for cls in torch.unique(support_y):
                        class_embeddings = model(support_x)[support_y == cls]
                        if len(class_embeddings) > 0:
                            prototypes.append(class_embeddings.mean(dim=0))
                    prototypes = torch.stack(prototypes) if len(prototypes) > 0 else torch.zeros((len(torch.unique(support_y)), model.embed_dim))

                    adv_logits_attack = -torch.cdist(adv_emb, prototypes)
                    total_loss += nn.CrossEntropyLoss()(adv_logits_attack, y_mapped)

                total_loss /= len(space_adv_embs_attack)  # Average loss across spaces

                # Compute gradient
                grad = torch.autograd.grad(total_loss, query_x_adv)[0]

                # Update adversarial examples with gradient sign
                query_x_adv = query_x_adv.detach() + model.pgd_step_size * grad.sign()

                # Project back to epsilon ball and valid data range
                delta = torch.clamp(query_x_adv - query_x, -attack_strength, attack_strength)
                query_x_adv = torch.clamp(query_x + delta, query_x.min(), query_x.max()).detach()

            # Evaluate adversarial examples
            with torch.no_grad():
                # Get projections from all spaces for adversarial examples
                space_adv_embs_eval = model(query_x_adv, return_all=True)

                # Combine predictions from all spaces
                combined_adv_logits = None

                for adv_emb in space_adv_embs_eval:
                    # Use the same prototypes from the clean support set for adversarial evaluation
                    prototypes = []
                    for cls in torch.unique(support_y):
                        class_embeddings = model(support_x)[support_y == cls]
                        if len(class_embeddings) > 0:
                            prototypes.append(class_embeddings.mean(dim=0))
                    prototypes = torch.stack(prototypes)

                    # Compute logits for this space
                    dists = torch.cdist(adv_emb, prototypes)
                    space_logits = -dists

                    if combined_adv_logits is None:
                        combined_adv_logits = space_logits
                    else:
                        combined_adv_logits += space_logits

                # Average logits across spaces
                combined_adv_logits /= len(space_adv_embs_eval)
                adv_probs = torch.softmax(combined_adv_logits, dim=1)
                adv_preds = combined_adv_logits.argmax(dim=1).cpu().numpy()
                adv_max_probs = adv_probs.max(dim=1)[0].cpu().numpy()

            # Map episode labels to global class indices and collect
            true_global_labels_episode = [selected_classes[label] for label in true_labels_episode]
            clean_pred_global_labels_episode = [selected_classes[label] for label in clean_preds]
            adv_pred_global_labels_episode = [selected_classes[label] for label in adv_preds]

            all_true_labels_global.extend(true_global_labels_episode)
            all_clean_pred_labels_global.extend(clean_pred_global_labels_episode)
            all_adv_pred_labels_global.extend(adv_pred_global_labels_episode)
            clean_confidences.extend(clean_max_probs)
            adv_confidences.extend(adv_max_probs)

            # Update confusion matrices
            for true, pred in zip(true_global_labels_episode, clean_pred_global_labels_episode):
                clean_conf_matrix[true, pred] += 1
            for true, pred in zip(true_global_labels_episode, adv_pred_global_labels_episode):
                adv_conf_matrix[true, pred] += 1

        except ValueError as e:
            print(f"Skipping episode: {e}")
            continue

    # Calculate metrics from confusion matrices
    clean_metrics = compute_metrics_from_confusion(clean_conf_matrix)
    adv_metrics = compute_metrics_from_confusion(adv_conf_matrix)

    # Calculate additional robustness metrics
    clean_correct = np.sum(np.diag(clean_conf_matrix))
    adv_correct = np.sum(np.diag(adv_conf_matrix))
    total_samples = np.sum(clean_conf_matrix)

    # Attack Success Rate (ASR)
    if clean_correct > 0:
        asr = (clean_correct - adv_correct) / clean_correct
    else:
        asr = 0.0

    # Robust Accuracy Ratio
    if clean_metrics['accuracy'] > 0:
        robust_ratio = adv_metrics['accuracy'] / clean_metrics['accuracy']
    else:
        robust_ratio = 0.0

    # Mean confidence drop
    mean_confidence_drop = np.mean(np.array(clean_confidences) - np.array(adv_confidences)) if clean_confidences else 0

    # Cohen's Kappa
    kappa_clean = cohen_kappa_score(all_true_labels_global, all_clean_pred_labels_global)
    kappa_adv = cohen_kappa_score(all_true_labels_global, all_adv_pred_labels_global)

    # Add additional metrics
    clean_metrics.update({
        'kappa': kappa_clean,
        'mean_confidence': np.mean(clean_confidences) if clean_confidences else 0
    })

    adv_metrics.update({
        'attack_success_rate': asr,
        'robust_accuracy_ratio': robust_ratio,
        'mean_confidence_drop': mean_confidence_drop,
        'kappa': kappa_adv,
        'mean_confidence': np.mean(adv_confidences) if adv_confidences else 0
    })

    return clean_metrics, adv_metrics, (clean_conf_matrix, adv_conf_matrix)

def plot_robustness_curves(results):
    """Plot clean vs adversarial accuracy across attack strengths with enhanced fonts"""
    # Filter results to get unique attack strengths and corresponding accuracies
    unique_strengths = sorted(list(set([r['attack_strength'] for r in results])))
    standard_clean_accs = []
    standard_adv_accs = []
    robust_clean_accs = []
    robust_adv_accs = []

    for strength in unique_strengths:
        # Get all scenarios for this attack strength and average them
        scenario_data = [r for r in results if r['attack_strength'] == strength]
        standard_clean_accs.append(np.mean([r['standard_clean_acc'] for r in scenario_data]))
        standard_adv_accs.append(np.mean([r['standard_adv_acc'] for r in scenario_data]))
        robust_clean_accs.append(np.mean([r['robust_clean_acc'] for r in scenario_data]))
        robust_adv_accs.append(np.mean([r['robust_adv_acc'] for r in scenario_data]))

    plt.figure(figsize=(10, 6))
    plt.plot(unique_strengths, standard_clean_accs, 'b-', label='Standard (Clean)', linewidth=2)
    plt.plot(unique_strengths, standard_adv_accs, 'b--', label='Standard (Adv)', linewidth=2)
    plt.plot(unique_strengths, robust_clean_accs, 'g-', label='Robust (Clean)', linewidth=2)
    plt.plot(unique_strengths, robust_adv_accs, 'g--', label='Robust (Adv)', linewidth=2)

    plt.xlabel('Attack Strength (ε)', fontsize=14)
    plt.ylabel('Accuracy', fontsize=14)
    plt.title('Model Robustness Across Attack Strengths', fontsize=16)
    plt.legend(fontsize=12)
    plt.grid(True)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.show()

def plot_performance_vs_shots(model, X, y, n_way, q_query, max_shots=20, num_trials=10, attack_strength=0.1):
    """Plot accuracy vs number of support shots for different few-shot scenarios"""
    shot_counts = range(1, max_shots+1)

    # We'll evaluate three scenarios: 1-shot, 5-shot, and 10-shot
    scenarios = [
        {'n_way': n_way, 'k_shot': 1, 'label': '1-shot', 'color': 'b', 'marker': 'o'},
        {'n_way': n_way, 'k_shot': 5, 'label': '5-shot', 'color': 'g', 'marker': 's'},
        {'n_way': n_way, 'k_shot': 10, 'label': '10-shot', 'color': 'r', 'marker': 'd'}
    ]

    plt.figure(figsize=(10, 6))

    for scenario in scenarios:
        accuracies = []
        for k in shot_counts:
            acc = 0
            for _ in range(num_trials):
                acc += evaluate_mspl(model, X, y, scenario['n_way'], k, q_query)
            accuracies.append(acc / num_trials)

        plt.plot(shot_counts, accuracies,
                 label=f"{scenario['label']}",
                 color=scenario['color'],
                 marker=scenario['marker'],
                 linewidth=1)

    plt.xlabel('Number of Support Shots (k)', fontsize=16)
    plt.ylabel('Accuracy', fontsize=16)
    plt.title(f'Few-Shot Performance vs Support Set Size (Attack Strength: {attack_strength})', fontsize=16)
    plt.legend(fontsize=14)
    plt.grid(True)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.show()

def print_detailed_metrics(metrics, class_names, title):
    """Print detailed metrics including per-class precision, recall, and F1"""
    print(f"\n{title} Metrics:")
    print(f"Accuracy: {metrics['accuracy']:.4f}")
    print(f"Weighted Precision: {metrics['weighted_precision']:.4f}")
    print(f"Weighted Recall: {metrics['weighted_recall']:.4f}")
    print(f"Weighted F1: {metrics['weighted_f1']:.4f}")

    if 'kappa' in metrics:
        print(f"Cohen's Kappa: {metrics['kappa']:.4f}")

    if 'attack_success_rate' in metrics:
        print(f"Attack Success Rate: {metrics['attack_success_rate']:.4f}")
        print(f"Robust Accuracy Ratio: {metrics['robust_accuracy_ratio']:.4f}")
        print(f"Mean Confidence Drop: {metrics['mean_confidence_drop']:.4f}")

# Modified scenario testing function to use MSPL with PGD
def run_adversarial_scenarios(X_train, y_train, X_test, y_test, class_names, num_classes, scenarios):
    results = []

    for scenario in scenarios:
        n_way = scenario['n_way']
        k_shot = scenario['k_shot']
        q_query = scenario.get('q_query', 15)
        attack_strength = scenario.get('attack_strength', 0.1)

        print(f"\nRunning scenario: {n_way}-way {k_shot}-shot with PGD attack strength {attack_strength}")

        # Split train into train and validation
        X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
            X_train, y_train, test_size=0.2, random_state=42)

        # Initialize standard model (no adversarial training)
        standard_model = MSPLNet(X_train.shape[1])
        standard_optimizer = optim.Adam(standard_model.parameters(), lr=0.001)

        # Train standard model
        print("Training standard MSPL model...")
        train_mspl(
            standard_model, standard_optimizer, X_train_split, y_train_split,
            X_val_split, y_val_split, n_way, k_shot, q_query, epochs=25,
            episodes_per_epoch=100, adversarial_training=False)

        # Initialize robust model (with adversarial training)
        robust_model = MSPLNet(X_train.shape[1])
        robust_optimizer = optim.Adam(robust_model.parameters(), lr=0.001)

        # Train robust model
        print("Training robust MSPL model with PGD adversarial training...")
        train_mspl(
            robust_model, robust_optimizer, X_train_split, y_train_split,
            X_val_split, y_val_split, n_way, k_shot, q_query, epochs=25,
            episodes_per_epoch=100, adversarial_training=True)

        # Evaluate both models under PGD attack with enhanced metrics
        print("Evaluating standard MSPL model under PGD attack...")
        standard_clean_metrics, standard_adv_metrics, standard_confs = enhanced_comprehensive_evaluate(
            standard_model, X_test, y_test, class_names, num_classes, n_way, k_shot, q_query,
            num_episodes=100, attack_strength=attack_strength)

        print("Evaluating robust MSPL model under PGD attack...")
        robust_clean_metrics, robust_adv_metrics, robust_confs = enhanced_comprehensive_evaluate(
            robust_model, X_test, y_test, class_names, num_classes, n_way, k_shot, q_query,
            num_episodes=100, attack_strength=attack_strength)

        # Print detailed metrics
        print_detailed_metrics(standard_clean_metrics, class_names, "Standard MSPL Model (Clean)")
        print_detailed_metrics(standard_adv_metrics, class_names, "Standard MSPL Model (Adversarial)")
        print_detailed_metrics(robust_clean_metrics, class_names, "Robust MSPL Model (Clean)")
        print_detailed_metrics(robust_adv_metrics, class_names, "Robust MSPL Model (Adversarial)")

        # Generate visualizations for robust model
        print("\nGenerating visualizations for robust model...")

        # Only plot performance vs support size for attack strength 0.1
        if attack_strength == 0.1:
            plot_performance_vs_shots(robust_model, X_test, y_test, n_way, q_query, attack_strength=attack_strength)

        # Store results
        scenario_result = {
            'n_way': n_way,
            'k_shot': k_shot,
            'attack_strength': attack_strength,
            'standard_clean_acc': standard_clean_metrics['accuracy'],
            'standard_adv_acc': standard_adv_metrics['accuracy'],
            'standard_drop': standard_clean_metrics['accuracy'] - standard_adv_metrics['accuracy'],
            'robust_clean_acc': robust_clean_metrics['accuracy'],
            'robust_adv_acc': robust_adv_metrics['accuracy'],
            'robust_drop': robust_clean_metrics['accuracy'] - robust_adv_metrics['accuracy'],
            'improvement': (robust_adv_metrics['accuracy'] - standard_adv_metrics['accuracy']) / standard_adv_metrics['accuracy'] if standard_adv_metrics['accuracy'] > 0 else 0,
            'standard_asr': standard_adv_metrics['attack_success_rate'],
            'robust_asr': robust_adv_metrics['attack_success_rate'],
        }
        results.append(scenario_result)

    # Plot robustness curves after all scenarios
    plot_robustness_curves(results)

    return results

# Main execution with enhanced evaluation
if __name__ == "__main__":
    # Load and preprocess data
    X_scaled, y_encoded, class_names, num_classes = load_and_preprocess_data(
        '/content/drive/MyDrive/ID2024-5G/Preprocessed_5G.csv')

    # Split into meta-train and meta-test sets
    train_mask, test_mask = split_meta_data(y_encoded, num_classes)
    X_train, y_train = X_scaled[train_mask], y_encoded[train_mask]
    X_test, y_test = X_scaled[test_mask], y_encoded[test_mask]

    # Define scenarios to test with different PGD attack strengths
    scenarios = [
        {'n_way': 5, 'k_shot': 1, 'attack_strength': 0.05},
        {'n_way': 5, 'k_shot': 5, 'attack_strength': 0.05},
        {'n_way': 5, 'k_shot': 10, 'attack_strength': 0.05},

        {'n_way': 5, 'k_shot': 1, 'attack_strength': 0.10},
        {'n_way': 5, 'k_shot': 5, 'attack_strength': 0.10},
        {'n_way': 5, 'k_shot': 10, 'attack_strength': 0.10},

        {'n_way': 5, 'k_shot': 1, 'attack_strength': 0.15},
        {'n_way': 5, 'k_shot': 5, 'attack_strength': 0.15},
        {'n_way': 5, 'k_shot': 10, 'attack_strength': 0.15},

        {'n_way': 5, 'k_shot': 1, 'attack_strength': 0.20},
        {'n_way': 5, 'k_shot': 5, 'attack_strength': 0.20},
        {'n_way': 5, 'k_shot': 10, 'attack_strength': 0.20},
    ]

    # Run enhanced adversarial scenarios with MSPL and PGD
    results = run_adversarial_scenarios(
        X_train, y_train, X_test, y_test, class_names, num_classes, scenarios)

    # Display results
    print("\nEnhanced Adversarial Scenario Results:")
    print(tabulate(results, headers="keys", tablefmt="grid", floatfmt=".4f"))
```
